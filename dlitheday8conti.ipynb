```python
import cv2 
```

```python
import numpy as np
#import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns
%matplotlib inline
```

```python
print("OpenCV-Python Version %s"%cv2.__version__)
```

```python
img=cv2.imread('messi.jpg',cv2.IMREAD_COLOR)
#if image path is wrong the resulting img will be null
if img is None:
    print('Open Error')
else:
    print('Image Loaded')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
print(type(img))
print(img.shape)
print(img.dtype)
print(img[:2,:2,:])
```

```python
img2=np.zeros((512,512,3),np.uint8)
plt.imshow(img2)
```

```python
cv2.line(img2,(0,0),(511,511),(255,0,0),5)
plt.imshow(img2)
```

```python
cv2.rectangle(img2,(384,0),(510,128),(0,255,0),3)
plt.imshow(img2)
```

```python
cv2.circle(img2,(447,63),63,(0,0,255),-1)
plt.imshow(img2)
```

```python
#ellipse(image,center,axes,angle,st)
cv2.ellipse(img2,(256,256),(100,50),-45,0,180,(255,0,0),-1)
plt.imshow(img2)
```
